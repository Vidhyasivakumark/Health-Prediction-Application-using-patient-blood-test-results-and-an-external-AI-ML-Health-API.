```python


import pandas as pd

df = pd.read_csv("final_data.csv")


print("Original Dataset:")
print(df.head())
# ==========================================

def detect_disease(row):

    diseases = []

    sugar = row['Sugar']
    hemoglobin = row['HemoglobinLevels']
    cholesterol = row['CholesterolTotal']

    # =====================================================
    # DIABETES CHECK
    # =====================================================

    diabetes_score = 0

    if sugar >= 6.5:
        diabetes_score += 1


    if diabetes_score >= 1:
        diseases.append("Diabetes")

    # =====================================================
    # HEART DISEASE RISK CHECK
    # =====================================================

    heart_score = 0

    if cholesterol >= 240:
        heart_score += 1

    if heart_score >= 1:
        diseases.append("Heart Disease Risk")

    # =====================================================
    # ANEMIA CHECK
    # =====================================================

    anemia_score = 0

    if hemoglobin < 12:
        anemia_score += 1

    if anemia_score >= 1:
        diseases.append("Anemia")

    # =====================================================
    # FINAL OUTPUT
    # =====================================================

    if len(diseases) == 0:
        return "Healthy"

    else:
        return " + ".join(diseases)

# =========================================================
# CREATE NEW DISEASE COLUMN
# =========================================================

df['disease_name'] = df.apply(detect_disease, axis=1)

# =========================================================
# DISPLAY UPDATED DATASET
# =========================================================

print("\nUpdated Dataset:")
print(df.head())

# =========================================================
# CLASS DISTRIBUTION
# =========================================================

print("\nDisease Distribution:")
print(df['disease_name'].value_counts())

# =========================================================
# SAVE UPDATED DATASET
# =========================================================

output_path = "/content/medical_disease_mapped.csv"

df.to_csv(output_path, index=False)

print("\nUpdated dataset saved successfully!")
print("File Location:", output_path)

```
